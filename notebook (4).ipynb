```python
# Import modules

import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
import requests
```

## Web Scraping

### Site: https://naijauto.com/market-news/innoson-motors-price-list-2018-288

```python
# Get data from site

request = requests.get("https://naijauto.com/market-news/innoson-motors-price-list-2018-288", "html.parser")
```

```python
# Create a soup object

soup = bs(request.text)
```

```python
# View the contents of soup object

soup.prettify()
```

```python
# Make a list to contain each car and it's price

soup.find_all("h3", style="")
```

```python
l = soup.find_all("h3", style="")

for i in range(len(l)):
    print(i, l[i].get_text())
```

```python
# Now, we can see clearly that our data ranges from 0 to 10

my_arr = np.arange(0, 10, 1)

name_list = []

for i in my_arr:
    name_list.append(l[i].get_text())
    
name_list
```

```python
lis = soup.find_all("ul")

for i in range(len(lis)):
    print(i, lis[i])
```

```python
# Looking at the above data, we get that our desired data ranges from 5 to 26

mainl = []

arr = np.arange(5, 27, 1)

for i in arr:
    mainl.append(lis[i].get_text())
    
mainl
```

```python
# Clean data

for i in range(len(name_list)):
    print(name_list[i])
```

```python
for i in range(len(mainl)):
    print(i, mainl[i])
```

```python
# Drop items 4, 10 and 17

new_list = mainl[:4] + mainl[4+1 :10] + mainl[10+1:17] + mainl[17+1 :]
```

```python
for i in range(len(new_list)):
    print(i, new_list[i])
```

```python
# Add missing data to name_list

name_list.insert(0, "1.  Innoson FOX: ₦3.52 - ₦3.85 million")
name_list.append("12.  Innoson 6800: ₦13.31 million")

name_list
```

```python
my_dict = {"Code name": ["Fox", "UMU", "G5", "G6", "G80", "G40", "4x4", "4x2", "5000", "6601", "6857", "6800"], "Name and Price": [name_list[0], name_list[1], name_list[2], name_list[3], name_list[4], name_list[5], name_list[6], name_list[7], name_list[8], name_list[9], name_list[10], name_list[11]], "Extra info": [[new_list[0], new_list[1]], [new_list[2], new_list[3]], [new_list[4], new_list[5]], [new_list[6]], [new_list[7]], [new_list[8]], [new_list[9], new_list[10]], [new_list[11], new_list[12]], [new_list[13], new_list[14]], [new_list[15]], [new_list[16]], [new_list[17], new_list[18]]]}
```

```python
# Convert to dataframe

df = pd.DataFrame(my_dict)
```

```python
df.head()
```

```python
# Data cleaning 

new = df["Name and Price"].str.split(": ", expand=True)

df["Name"] = new[0]
df["Price"] = new[1]
```

```python
df.head()
```

```python
new = df["Name"].str.split(".", expand=True)

df["Index"] = new[0]
df["Name"] = new[1]
```

```python
df = df.set_index("Index")
df.head()
```

```python
df = df.drop(["Code name", "Name and Price"], axis=1)
```

```python
df.head()
```

I will proceed with using the name and price for analysis

```python
data = df[["Name", "Price"]]
```

```python
# Clean the price column

tab = data["Price"].str.split("-", expand = True)
```

```python
data["Min price"] = tab[0]
data["Max price"] = tab[1]
```

```python
data
```

```python
data = data.rename(columns={'Price': 'Price range'})
```

```python
data["Min price"] = data["Min price"].str.strip("₦")
data["Min price"] = data["Min price"].str.replace(",", "")
data["Min price"] = data["Min price"].str.replace(" million", "")
```

```python
data
```

```python
data["Max price"] = data["Max price"].str.replace("₦", "")
```

```python
data["Max price"] = data["Max price"].str.replace("million", "")
```

```python
data["Max price"] = data["Max price"].str.replace(" million", "")
```

```python
data
```

```python
data['Min price'] = data['Min price'].astype(float)
data['Max price'] = data['Max price'].astype(float)
```

```python
data
```

```python
data[data['Min price'] > 100]['Min price']/1000000
```

```python
data
```

```python
data.iloc[5, 2] = data.iloc[5, 2] / 1000000
```

```python
data.iloc[4, 2] = data.iloc[4, 2] / 1000000
```

```python
data
```

```python
# Create a new column for the average price

data['Average price'] = data['Min price']
```

```python
data
```

```python
data.iloc[0, 4] = (data.iloc[0, 2] + data.iloc[0, 3]) / 2
data.iloc[1, 4] = (data.iloc[1, 2] + data.iloc[1, 3]) / 2
data.iloc[2, 4] = (data.iloc[2, 2] + data.iloc[2, 3]) / 2
data.iloc[3, 4] = (data.iloc[3, 2] + data.iloc[3, 3]) / 2
data.iloc[8, 4] = (data.iloc[8, 2] + data.iloc[8, 3]) / 2
```

```python
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
figure = sns.barplot(x='Name', y='Average price', data=data)
plt.xticks(rotation=90)
plt.xlabel("Model name")
plt.ylabel("Average price")
plt.title("Models versus price bar plot")
plt.show()
```

### Innoson G80 has the highest price while Innoson FOX has the least price

```python
figure = sns.lineplot(x='Name', y='Min price', data=data)
plt.xticks(rotation=90)
plt.xlabel("Model name")
plt.ylabel("Minimum price")
plt.title("Models versus minimum price line plot")
plt.show()
```
